```python
from ultralytics import YOLO

# Cargar el modelo YOLO preentrenado
model = YOLO('/content/Yolo/yolo11s.pt')

# Ruta del archivo de configuraciÃ³n YAML
data_yaml = "/content/Yolo/dataset/data.yaml"  # AsegÃºrate de que el archivo existe y tiene la estructura correcta

# Entrenar el modelo
model.train(
    data=data_yaml,
    epochs=50,
    imgsz=640,
)
```
