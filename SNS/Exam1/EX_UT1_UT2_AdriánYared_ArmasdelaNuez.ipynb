## 1.a) Genera datos aleatorios que representen el número de pasajeros en tres ferrys. Los barcos son: Ferry A, Ferry B y Ferry C

Cada barco tiene 100 pasajeros generalos aleatoriamente con una distribución normal, utilizando las siguientes especificaciones.
- Ferry A: Media (η\mu)=200 pasajeros, Desviación estandar (o\sigma) = 20
- Ferry B: Media (η\mu)=300 pasajeros, Desviación estandar (o\sigma) = 30
- Ferry C: Media (η\mu)=250 pasajeros, Desviación estandar (o\sigma) = 25

Representa cada conjunto de datos en histogramas separados (uno por cada ferry). cada histograma debe incluir:
- Un título ondicando el ferry representado ("Histograma Ferry A").
- Etiqueta de los ejes correctamente definidas.
- El número de barras/columnas a mostrar en cada histograma ha de ser 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar de datos aleatorios
np.random.seed(60) # Con esto aseguro que son compleamente aleatorios

# Datos de los ferrys del enunciado
mu_A, sigma_A = 200, 20
mu_B, sigma_B = 300, 30
mu_C, sigma_C = 250, 25

# Datos para cada ferry
data_A = np.random.normal(mu_A, sigma_A, 100)
data_B = np.random.normal(mu_B, sigma_B, 100)
data_C = np.random.normal(mu_C, sigma_C, 100)

# Función del gráfico para histogramas
def plot_histogram(data, title, xlabel, ylabel): # imprime el titulo (title), etiquetas en los ejes (xlabel, ylabel)
    plt.hist(data, bins=10, color='skyblue', edgecolor='black') # bins indica el n de barras en cada uno (10)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Graficamos cada histograma (Title, xlabel, ylabel en cada uno respectivamente)
plot_histogram(data_A, "Histograma Ferry A", "Número de pasajeros", "Frecuencia")
plot_histogram(data_B, "Histograma Ferry B", "Número de pasajeros", "Frecuencia")
plot_histogram(data_C, "Histograma Ferry C", "Número de pasajeros", "Frecuencia")
```

```python
import pandas as pd

# Crear un DataFrame con los datos generados anteriormente
data = {
    "Ferry A": data_A,
    "Ferry B": data_B,
    "Ferry C": data_C
}
df_ferries = pd.DataFrame(data)

# Resmen
resumen_estadistico = df_ferries.describe().T  # Transpuesta

# N muestras
resumen_estadistico['Número de filas'] = df_ferries.count().values

# Mostrar el DataFrame
resumen_estadistico[['Número de filas', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]
```

#3)Estamos analizando un conjunto de datos aleatorios y deseamos identificar los outliers para lo cual es necesario generar 50 números aleatorios con la distribución normal, con media 50 y desviación estándar 15. La variable donde se guarden se ha de llamar "datosEstudio" y en la posición 10, 20 y 30 guardar los valores 105, 200 y 300 respectivamente. Para identificar los outliers se han de utilizar los siguientes métodos:

##3.a) Utilizando la Técnica basada en la probabilidad: Aplica/programa una solución asumiendo que los datos siguen una distribucuión normal para calcular la probabilidad de ocurrencia de cada observación. Identifica los valores cuya probabilidad sea inferior a 0.01% para considerarlos outliers.

image.png
<br>
image.png

```python
import numpy as np
from scipy.stats import norm

#  Datos
np.random.seed(42)
datosEstudio = np.random.normal(50, 15, 50)

# pos 10, 20, y 30
datosEstudio[9] = 105
datosEstudio[19] = 200
datosEstudio[29] = 300

# prob distribución normal
probabilidades = norm.pdf(datosEstudio, 50, 15)

# outliers menor a 0.01
outliers = datosEstudio[probabilidades < 0.0001]
outliers
```

##3.b) Utilizando la Técnica basada en los cuartiles: Calcular/programa el rando intercuartil (IQR) y utiliza las siguientes reglas para identificar los valores atípicos:

###3.b.1) Para los outliers leves (mild outlier): si está fuera de rango:
- [Q1-1.5x|QR,Q3+1.5xIQR]

```python
import numpy as np

# datos
np.random.seed(42)  #random
datosEstudio = np.random.normal(50, 15, 50) #normal

# cambio valores pedidos
datosEstudio[9] = 105
datosEstudio[19] = 200
datosEstudio[29] = 300

# quartiles
Q1 = np.percentile(datosEstudio, 25)
Q3 = np.percentile(datosEstudio, 75)

# rango intercuartil (IQR)
IQR = Q3 - Q1

# ímites outliers leves
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# outliers leves
outliers_leves = datosEstudio[(datosEstudio < limite_inferior) | (datosEstudio > limite_superior)]
outliers_leves
```

3.b.2) Para los outliers extremos (extreme outlier): si está fuera de rango:
- [Q1-3x|QR,Q3+3x|QR]

```python
import numpy as np

# datos
np.random.seed(42)  #random
datosEstudio = np.random.normal(50, 15, 50) #normal

# cambio valores pedidos
datosEstudio[9] = 105
datosEstudio[19] = 200
datosEstudio[29] = 300

# quartiles
Q1 = np.percentile(datosEstudio, 25)
Q3 = np.percentile(datosEstudio, 75)

# rango intercuartil (IQR)
IQR = Q3 - Q1

# ímites outliers leves
limite_inferior = Q1 - 3 * IQR
limite_superior = Q3 + 3 * IQR

# outliers leves
outliers_extremo = datosEstudio[(datosEstudio < limite_inferior) | (datosEstudio > limite_superior)]
outliers_extremo
```

Las técnicas de probabilidad y cuartiles ofrecen diferentes enfoques para analizar sus correspondientes distribuciones. La probabilidad se enfoca en la predicción dentro de una muestra, utilizando una función que asigna probabilidades a cada resultado posible. Y  los cuartiles dividen los datos en segmentos (quarttiles), proporcionando una visión de la dispersión y los valores extremos. Mientras que la probabilidad se centra en la estimación de eventos, los cuartiles ofrecen una descripción de la distribución de los datos.
